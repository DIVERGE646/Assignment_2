```python
#Describe the role of predefined keywords in Python and provide examples of how they are used in a
program

#ANSWER:


In Python, predefined keywords (also known as reserved words) are words that have a specific meaning in the
language's syntax. These keywords are part of Python's core language, and they define the structure and behavior
of the code. Because of their specific functionality, they cannot be used as identifiers (i.e., variable
function, or class names).

Python keywords are case-sensitive and help define the operations, control structures, and data types
in a program.

Role of Predefined Keywords
Syntax Definition: Keywords are the building blocks of Python syntax, helping structure the flow of the
program (e.g., if, else, while, for).
Control Flow: They control how the program is executed, whether conditionally (if, else, elif), repeatedly
(while, for), or in an error-handling context (try, except, finally).
Variable and Function Declaration: Keywords like def, class, and lambda are used to define functions, classes,
and anonymous functions (lambdas).
Memory Management: Some keywords are related to memory management (e.g., del to delete an object).
Logical Operations: Keywords like and, or, and not are used in logical expressions.
```

```python
#Compare and contrast mutable and immutable objects in Python with examples

#ANSWER:

#Comparing Mutable vs Immutable Objects in Practice
1. Memory Efficiency
Mutable objects like lists are more memory efficient when frequent modifications are needed since they are
updated in place without creating new objects.
Immutable objects like strings, however, may use more memory because every modification results in the creation
of a new object.
2. Thread Safety
Immutable objects are generally safer to use in multithreaded environments because their state cannot be
changed once created. This prevents race conditions.
Mutable objects can lead to issues in multithreaded programs if one thread modifies an object while another
is using it.
3. Use in Data Structures
Immutable objects (like tuples) are often used as dictionary keys or elements of sets because their
contents remain constant, ensuring reliable hashing.
Mutable objects (like lists or dictionaries) cannot be used as dictionary keys or set elements because
their contents can change, which would break the hash-based data structure.
```

```python
#Explain the concept of type casting in Python with examples

#ANSWER:

Type casting (also known as type conversion) in Python refers to converting one data type into another. This is
often required when different types of variables need to be used together in operations. Python provides several built-in functions to perform type casting, which can be either implicit or explicit.

1. Implicit Type Casting
In implicit type casting, Python automatically converts one data type to another without any explicit
instruction from the programmer. This usually happens when performing operations between different data types
where Python upgrades the lower precision type to a higher precision type

2.Explicit type casting:
In Python refers to manually converting a variable from one data type to another using built-in functions.
Unlike implicit type casting, where Python automatically handles the conversion, explicit type casting
requires the programmer to indicate the desired conversion.

Common Functions for Explicit Type Casting
Here are some of the most frequently used functions for type casting in Python:

int(): Converts to an integer.

Syntax: int(value)
Converts numeric strings or float values to integers (truncating the decimal part).
float(): Converts to a floating-point number.

Syntax: float(value)
Converts integers or numeric strings to floats.
str(): Converts to a string.

Syntax: str(value)
Converts numbers, lists, or other data types to strings.
list(): Converts to a list.

Syntax: list(iterable)
Converts sequences like strings, tuples, or sets into lists.
tuple(): Converts to a tuple.

Syntax: tuple(iterable)
Converts sequences like lists, strings, or sets into tuples.
```

```python
#How do conditional statements work in Python? Illustrate with examples

#ANSWER:

Conditional statements in Python are used to execute certain blocks of code based on specific conditions.
These conditions evaluate to either True or False, and based on that result, the appropriate block of code is
executed.

Types of Conditional Statements in Python:
if Statement
if-else Statement
if-elif-else Ladder
Nested if Statements

1. if Statement
The if statement is the most basic form of a conditional statement. It checks whether a condition is True.
If the condition evaluates to True, the indented block of code below the if statement is executed.

Syntax:
if condition:
 # code to execute if condition is True

Example:
python
x = 10
if x > 5:
    print("x is greater than 5")  # This will be executed because the condition is True
```
